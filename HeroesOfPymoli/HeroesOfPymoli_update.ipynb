## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Make dataframe for age and rename columns
age_df = purchase_df["Age"].value_counts()
age_df = pd.DataFrame(age_df).reset_index()
age_rename = age_df.rename(columns={'index': 'Age', 'Age': 'Total Count'})

# Add a variable for total players
total_plyrs = 576

# Add column for percentage of players in age and calculate it.
age_rename['Percentage of Players']= age_rename['Total Count']*100/total_plyrs

# Make a bin and label for age
age_bins = [0, 10, 14, 19, 24, 29, 34, 39, 45]
age_rng_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice data to place it into bins
pd.cut(age_rename["Age"], age_bins, labels=age_rng_labels)
age_rename['Age Range'] = pd.cut(age_rename["Age"], age_bins, labels=age_rng_labels)

# Groupby Age Range. Get total of players by count and percentage in each age group.
age_rename = age_rename.groupby('Age Range')
age_rename[['Age Range', 'Total Count', 'Percentage of Players']].sum()
```
